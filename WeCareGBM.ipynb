```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/content/Maternal Health Risk Data Set.csv")
```

```python
df.head()
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
print(X)
```

```python
print(y)
```

```python
label_map = {'low risk': 0, 'mid risk': 1, 'high risk': 2}
y_numeric = np.array([label_map[val] for val in y])
print(y_numeric)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
column_means = np.nanmean(X_train, axis=0)
X_train_filled = np.where(np.isnan(X_train), column_means, X_train)
X_test_filled = np.where(np.isnan(X_test), column_means, X_test)
print (X_test)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_filled)
X_test_scaled = scaler.transform(X_test_filled)

# Fitting Gradient Boosting to the Training set
classifier = GradientBoostingClassifier(n_estimators=600, learning_rate=0.1, max_depth=3, random_state=0)
classifier.fit(X_train_scaled, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test_scaled)
print(y_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
